# Apply

Signature: df.apply(func, axis=0, broadcast=None, raw=False, reduce=None, result_type=None, args=(), **kwds) Docstring: Apply a function along an axis of the DataFrame.
Objects passed to the function are Series objects whose index is either the DataFrame's index (axis=0) or the DataFrame's columns (axis=1). By default (result_type=None), the final return type is inferred from the return type of the applied function. Otherwise, it depends on the result_type argument.

```python
lst = []
for i in df['Meal Price']:
    lst.append(i*i)
```

```python
lst
```

```python
df['Meal Squared'] = list[lst]
```

```python
df['Meal Price'] = df['Meal Price'].astype('int')
df.drop("70000",axis = 0)
```

```python
df['Meal Squared apply'] = df['Meal Price'].apply(lambda x : x*x)
```

```python
df.loc[60000,'Meal Squared'] = "A String value"
```

```python
df['Meal Squared'].apply(lambda x: type(x)).value_counts()
```

```python
df['Meals Squared']['criteria to access rows']
```

```python
df['Meals Squared'][df['Meal Squared'].apply(lambda x: isinstance(x, str))].index
```

```python
df['Meal Squared'] = pd.to_numeric(df['Meal Squared'], errors = 'coerce')
```

```python
df.tail()
```

```python
df['Meal Squared'] = df['Meal Squared'].fillna(df['Meal Squared'].mean())
```

```python
df['Meal Squared apply'] = df['Meal Squared apply'].fillna(df['Meal Squared apply'].median())
```

```python
df['Meal Squared'].astype(int)
```

```python
df['Meal Squared func'] = df['Meal Squared func'].fillna(0)
df['Meal Squared func'] = df['Meal Squared func'].replace(0,df['Meal Squared apply'].median())
```
